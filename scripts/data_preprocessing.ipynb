```python
import pandas as pd
import os
import kagglehub
```

```python
# Path to kaggle dataset
path = kagglehub.dataset_download("vagifa/ethereum-frauddetection-dataset")

# List all files in the dataset directory
dataset_files = os.listdir(path)

# Download the file
csv_file = os.path.join(path, 'transaction_dataset.csv')
```

```python
df = pd.read_csv(csv_file)
df = df.drop(columns=['Unnamed: 0', 'Index'])
df
```

```python
df.columns
df.dtypes
```

```python
# remove the whitespace at the start of some columns
df.columns = df.columns.str.strip()
df.columns
```

```python
colsToBeDeleted = []
for col in df.select_dtypes(include="number").columns:
    if df[col].var() == 0:  # Check if the variance of the column is 0
        colsToBeDeleted.append(col)

print(colsToBeDeleted)

# drop features with constant variance
df = df.drop(columns = colsToBeDeleted)
```

```python
df.columns
df.dtypes
```

```python
# Dealing with missing values

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include='number').columns
categorical_cols = df.select_dtypes(include=['object']).columns

from sklearn.impute import KNNImputer

# Apply KNN Imputer to numerical columns
numerical_data = df[numerical_cols]
imputer = KNNImputer(n_neighbors=5)
numerical_data_imputed = pd.DataFrame(imputer.fit_transform(numerical_data), columns=numerical_cols)

# Impute categorical columns with the mode (maybe theres a btr thing to replace the missing values)
categorical_data = df[categorical_cols]
for col in categorical_cols:
    categorical_data[col] = categorical_data[col].fillna(categorical_data[col].mode()[0])


# Combine numerical and categorical data
df = pd.concat([numerical_data_imputed, categorical_data], axis=1)
print(df.isnull().sum())
```

```python
# drop the address column. or hash the address to numerics. idk which btr
#df = df.drop(columns=['Address'])

import hashlib

# Hash the Categorical columns
df['address hash'] = df['Address'].apply(lambda x: int(hashlib.sha256(x.encode('utf-8')).hexdigest(), 16))

df['ERC20 most sent token type hash'] = df['ERC20 most sent token type'].apply(
    lambda x: int(hashlib.sha256(x.encode('utf-8')).hexdigest(), 16) % 100
)

df['ERC20 most rec token type hash'] = df['ERC20_most_rec_token_type'].apply(
    lambda x: int(hashlib.sha256(x.encode('utf-8')).hexdigest(), 16) % 100
)

# ensure all hashed columns are numeric
df['address hash'] = pd.to_numeric(df['address hash'])
df['ERC20 most sent token type hash'] = pd.to_numeric(df['ERC20 most sent token type hash'])
df['ERC20 most rec token type hash'] = pd.to_numeric(df['ERC20 most rec token type hash'])

# Delete the original columns
df = df.drop(columns=['Address', 'ERC20 most sent token type', 'ERC20_most_rec_token_type'])
```

```python
#upsample -> increase fraud number to match non-fraud number
from sklearn.utils import resample
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X = df.drop(columns=['FLAG'])
y = df['FLAG']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Combine the X_train and y_train data back together to facilitate resampling
df_train = pd.concat([X_train, y_train], axis=1)

# Separate the majority and minority classes
df_majority = df_train[df_train['FLAG'] == 0]
df_minority = df_train[df_train['FLAG'] == 1]

# Upsample the minority class (fraud)
df_minority_upsampled = resample(df_minority,
                                 replace=True,    # Sample with replacement, since some instances frm minority class selected multiple times
                                 n_samples=len(df_majority),  # Match the majority class size
                                 random_state=42)  

# Combine the upsampled minority class with the majority class
df_balanced = pd.concat([df_majority, df_minority_upsampled])

# Shuffle the dataset to ensure randomness
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Split back into X and y
X_train_balanced = df_balanced.drop(columns=['FLAG'])
y_train_balanced = df_balanced['FLAG']
```

```python
#undersample -> decrease non-fraud number to match fraud number
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
X = df.drop(columns=['FLAG'])
y = df['FLAG']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Combine the X_train and y_train data back together to facilitate resampling
df_train = pd.concat([X_train, y_train], axis=1)

# Separate the majority and minority classes
df_majority = df_train[df_train['FLAG'] == 0]
df_minority = df_train[df_train['FLAG'] == 1]

# Downsample the majority class (non-fraud)
df_majority_downsampled = resample(df_majority,
                                   replace=False,    # Sample without replacement
                                   n_samples=len(df_minority),  # Match the minority class size
                                   random_state=42)  

# Combine the downsampled majority class with the minority class
df_balanced = pd.concat([df_majority_downsampled, df_minority])

# Shuffle the dataset to ensure randomness
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Split back into X and y
X_train_balanced = df_balanced.drop(columns=['FLAG'])
y_train_balanced = df_balanced['FLAG']
```
